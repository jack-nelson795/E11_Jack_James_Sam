```python
# this will be the slice for the new outdoor data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress


url_outdoors_2 = "https://raw.githubusercontent.com/jack-nelson795/E11_Jack_James_Sam/refs/heads/main/outdoor_data_2.csv"
df_out_2 = pd.read_csv(url_outdoors_2)

for i in range(0,900):
    df_out_2 = df_out_2.drop(index = i)

line = linregress(df_out_2['Tempreature'], df_out_2['Humidity'])
slope = line.slope
inter = line.intercept
R_2 = line.rvalue ** 2
print(f'The R squared value for this fit is: {R_2}')

line_fit = (slope * df_out_2['Tempreature']) + inter

plt.figure(1)

plt.scatter(df_out_2['Tempreature'], df_out_2['Humidity'])
plt.plot(df_out_2['Tempreature'], line_fit, color = 'red')
plt.title('Humidity vs Temp')
plt.xlabel('Temp')
plt.ylabel('Humidity')
plt.grid()
plt.show()

plt.figure(2)

plt.scatter(df_out_2['Tempreature'], df_out_2['Pressure'])
plt.title('Pressure vs Temp')
plt.xlabel('Temp')
plt.ylabel('Pressure')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
indoor_df = pd.read_csv('air_weather_indoors.csv')
outdoor_df = pd.read_csv('air_weather_outdoors.csv')

indoor_df = indoor_df.drop(index=0)
outdoor_df = outdoor_df.drop(index=0)
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Tempreature'])
plt.xlabel('Time (s)')
plt.ylabel('Temp. (degrees C)')
plt.title('Indoor Temp. vs Time')
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Gas'])
plt.xlabel('Time (s)')
plt.ylabel('Gas')
plt.title('Indoor Gas vs Time')
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Humidity'])
plt.xlabel('Time (s)')
plt.ylabel('Humidity')
plt.title('Indoor Humidity vs Time')
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Pressure'])
plt.xlabel('Time (s)')
plt.ylabel('Pressure')
plt.title('Indoor Pressure vs Time')
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Altitude'])
plt.xlabel('Time (s)')
plt.ylabel('Altitude')
plt.title('Indoor Altitude vs Time')
```

```python
plt.plot(indoor_df['t_s'],indoor_df['Pressure'])
plt.xlabel('Time (s)')
plt.ylabel('Pressure')
plt.title('Indoor Pressure vs Time')
```

```python
for i in range(1,len(indoor_df.columns)):
    plt.plot(indoor_df['t_s'],indoor_df.columns[i])
    plt.xlabel('Time (s)')
    plt.ylabel(indoor_df.columns[i])
    plt.title(f'Indoor {indoor_df.columns[i]} vs Time')
```
